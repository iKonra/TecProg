Ejercicio 2 Conversión de Segundos (*)
Realiza un programa que convierta un número de segundos introducidos por el usuario en horas,
minutos y segundos

```python
cant_segundos = int(input("Introduce una cantidad en segundos: "))
horas = cant_segundos//3600
resto_hora = cant_segundos%3600
minutos = resto_hora//60
segundos = resto_hora%60

print(f"horas  {horas} minutos {minutos} segundos {segundos}" )
```

Ejercicio 3
Calculadora de Raíces Cuadradas
Crea un programa que calcule y muestre la raíz cuadrada de una serie de números introducidos por
el usuario. El programa termina cuando el usuario introduce un número negativo.


```python
import math
number = float(input("Ingresa un numero: "))

while number >= 0:
    print(math.sqrt(number))
    number = float(input("Ingresa un numero: "))
print(number)
```

Ejercicio 4 Clasificación de Consumo Eléctrico (*)
Crea un programa que solicite la cantidad de kilovatios hora (kWh) consumidos en un mes y calcule
el costo total del servicio según la siguiente tabla:

● Bajo. Hasta 100 kWh → $200 por kWh
● Medio. Entre 101 y 300 kWh → $300 por kWh
● Alto. Más de 300 kWh → $400 por kWh

```python
kwh = int(input("Cuantos kilovatios se consumio este mes: "))
if kwh <= 100: print(kwh*200)
if kwh > 100 and kwh <= 300: print(kwh*300)
if kwh > 300: print(kwh*400)
```

Ejercicio 1. Conteo hasta N (*)
Escribe un programa que pida al usuario un número N y luego cuente desde 1 hasta N.


```python
N = int(input("Ingrese un numero de 1 hasta N"))
for i in range(N):
    print(i)
```

Ejercicio 3. Media de Números (*)
Crea un programa que lea N números del usuario y calcule su media

```python
N = int(input("Cuantas entradas habra? "))
suma = 0
for i in range(N):
    numero = float(input("Ingrese un numero: "))
    suma += numero
media = suma/N
media

```

Ejercicio 2. Suma de Números (*)
Desarrolla un programa que solicite al usuario N números y calcule su suma.

```python
N = int(input("Cuantas entradas habra"))
suma = 0

for i in range(N):
    suma += int(input())
suma
```
